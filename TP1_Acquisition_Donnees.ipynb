### 2.3 Web Scraping
Extraction de données depuis le code HTML d'une page.
**Attention :** Vérifiez toujours le fichier `robots.txt` du site.

**Éthique :**
* Ne surchargez pas les serveurs.
* Ne volez pas de données personnelles.
* Respectez les droits d'auteur.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Site d'entrainement au scraping (autorisé)
url = "http://quotes.toscrape.com/"
response = requests.get(url)

# Création de la "soupe" (parseur HTML)
soup = BeautifulSoup(response.content, 'html.parser')

# Trouver toutes les citations et auteurs
quotes = soup.find_all('span', class_='text')
authors = soup.find_all('small', class_='author')

# Stocker dans une liste
quote_list = []

for quote, author in zip(quotes, authors):
    quote_list.append({
        'citation': quote.text,
        'auteur': author.text
    })

# Convertir en DataFrame
df_quotes = pd.DataFrame(quote_list)
print("--- Résultat du Scraping ---")
display(df_quotes)
```

### 2.4 Connexions aux bases de données (SQLite)
SQLite est une base de données légère incluse dans Python. Nous allons créer une DB, y écrire des données, puis les lire avec du SQL.

```python
import sqlite3
import pandas as pd

# 1. Créer une connexion (crée le fichier s'il n'existe pas)
conn = sqlite3.connect('students.db')

# 2. Créer des données exemples
data = {
    'student_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'major': ['CS', 'Math', 'CS', 'Physics', 'Math'],
    'gpa': [3.8, 3.6, 3.9, 3.7, 3.5]
}
df_students = pd.DataFrame(data)

# 3. Écrire le DataFrame dans la base SQL
df_students.to_sql('students', conn, if_exists='replace', index=False)
print("Données écrites dans la base SQLite 'students.db'.")

# 4. Interroger la base avec du SQL
# Exemple : Trouver les étudiants avec un GPA > 3.7
query = "SELECT * FROM students WHERE gpa > 3.7"
result = pd.read_sql_query(query, conn)

print("\n--- Résultat de la requête SQL (GPA > 3.7) ---")
display(result)

# 5. Fermer la connexion
conn.close()
```
